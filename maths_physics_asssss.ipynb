#### 2.b: Print out the total credits and calculte CPI of each student. 
CPI is the weighted average of core courses, and electives (weights being the letter grades converted to number AP,AA=10, AB=9, BB=9, BC=7, CC=6)

```python
import csv

filename = 'student_records.csv'


grade_to_points = {"AP": 10, "AA": 10, "AB": 9, "BB": 9, "BC": 7, "CC": 6}

student_data = {}

with open(filename, 'r') as csvfile:
    csvreader = csv.DictReader(csvfile)
    for row in csvreader:
        roll_no = row['roll_number']
        grade = row['grade']
        credit = int(row['credit'])

       
        if roll_no not in student_data:
            student_data[roll_no] = {'total_credits': 0, 'weighted_points': 0}

        
        student_data[roll_no]['total_credits'] += credit
        student_data[roll_no]['weighted_points'] += grade_to_points[grade] * credit


print("RollNO. Credits CPI")
for roll_no, data in student_data.items():
    total_credits = data['total_credits']
    weighted_points = data['weighted_points']
    cpi = weighted_points / total_credits if total_credits > 0 else 0
    print(f"{roll_no}  {total_credits}  {cpi:.2f}")
```

#### 2.c: Print out the names of all students who meet the graduation requirements 
Atleast 20 credist of core course, 15 credits of department elective, 10 credits of flexible elective and 5 credits of hasmed electives

```python
import csv


filename = "student_records.csv"
student_data = {}

with open(filename, 'r') as csvfile:
    csvreader = csv.DictReader(csvfile)

    
    for row in csvreader:
        roll_no = row['roll_number']
        course_type = row['course_type']
        credit = int(row['credit'])

        
        if roll_no not in student_data:
            student_data[roll_no] = {'core': 0, 'department_elective': 0, 
                                     'felxible_elective': 0, 'hasmed_elective': 0}

        if course_type in student_data[roll_no]:
            student_data[roll_no][course_type] += credit


qualified_students = []

for roll_no, credits in student_data.items():
    if (credits['core'] >= 20 and
        credits['department_elective'] >= 15 and
        credits['felxible_elective'] >= 10 and
        credits['hasmed_elective'] >= 5):

        qualified_students.append(roll_no)
        

for roll_no in qualified_students:
    print(roll_no)


```

#### 2.d: Print out the names of all students who completed a minor 
Atleast 10 credits with minor tag in a specific department

```python
import csv

target_department = "MA"

s_data = {}

with open("student_records.csv", 'r') as csvfile:
    csvreader = csv.DictReader(csvfile)

    for row in csvreader:
        roll_no = row['roll_number']
        department = row['department']
        course_type = row['course_type'] 
        credit = int(row['credit'])

       
        if roll_no not in s_data:
            s_data[roll_no] = {}
        if department not in s_data[roll_no]:
            s_data[roll_no][department] = 0

   
        if course_type == 'minor':
            s_data[roll_no][department] += credit

q_students = [
    roll_no for roll_no, departments in s_data.items()
    if departments.get(target_department, 0) >= 10
]

print("\n".join(q_students))
```

#### 2.e: Print out the names of all students who completed a honours

```python
import csv


filename = 'student_records.csv'
student_data = {}

with open(filename, 'r') as csvfile:
    csvreader = csv.DictReader(csvfile)
 
    for row in csvreader:
        roll_no = row['roll_number']
        tag = row['course_type']  
        credit = int(row['credit'])
        
        
        if roll_no not in student_data:
            student_data[roll_no] = {'honours_credits': 0, 'core_credits': 0}
        
        
        if tag == 'honours':
            student_data[roll_no]['honours_credits'] += credit
        elif tag == 'core':
            student_data[roll_no]['core_credits'] += credit


qualified_students = []

for roll_no, data in student_data.items():
    if data['honours_credits'] >= 10 and data['core_credits'] >= 20:
        qualified_students.append(roll_no)


for roll_no in qualified_students:
    print(roll_no)
```

## Part A
Minimize the function $f(x, y) = 2(x - y - 3)^2 + 4(x + 2y + 1)^4$.<br>
With the constraints : $ x - y \ge -3, (x + 2)^2 + (y + 1)^2 \le 5$ <br>
Using scipy.optimize.minimize with constraints. (Hint: Look at the examples in the official documentation)

```python
#TODO
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.optimize import minimize 


def fun(variable):
    x,y=variable
    return ((2*(x-y-3)**2)+(4*(x+2*y+1)**4))


def c1(variable):
    x,y=variable
    return (x-y+3)

def c2(variable):
    x,y=variable
    return (5-((x+2)**2)+((y+1)**2))

guess=[0,0]
 
constr=[{"type":"ineq","fun":c1},{"type":"ineq","fun":c2}]      #ineq= inequality constraint, eq=equality constraint
# print(constraints)

solution=minimize(fun,guess,method="SLSQP",constraints=constr)
print(solution.x)
print(solution.fun)
```

```python
### TODO 1
### Load data from data_path
### Check the input file spice_locations.txt to understand the Data Format
### Return : np array of size Nx2
def load_data(data_path):
    data=np.loadtxt(data_path,delimiter=',')
    return data
data=load_data("spice_locations.txt")          #here data is my numpy array of 60*2. 
data.shape
```

```python
### TODO 2.1
### If init_centers is None, initialize the centers by selecting K data points at random without replacement
### Else, use the centers provided in init_centers
### Return : np array of size Kx2
def initialise_centers(data, K, init_centers=None):
    if init_centers == None: 
        x=np.random.choice(data.shape[0],size=K,replace=False)       #np.random.choice is used to randomnly select indices
        y=data[x]
    else: 
        y=init_centers
    return y

centers=initialise_centers(data,4)
print(centers)
```

```python
### TODO 2.2
### Initialize the labels to all ones to size (N,) where N is the number of data points
### Return : np array of size N
def initialise_labels(data):
    x=data.shape[0]
    label=np.ones(x,dtype=int)
    return label
label=initialise_labels(data)
print(label)
label.shape[0]
```

```python
### TODO 3.1 : E step
### For Each data point, find the distance to each center
### Return : np array of size NxK
def calculate_distances(data, centers):
    x=data[:,None,:]-centers[None,:,:]       #broadcasting
    y = np.sum(x ** 2, axis=2)
    z=np.sqrt(y)
    
    return z
distance=calculate_distances(data,centers)
print(distance)
```

```python
### TODO 3.2 : E step
### For Each data point, assign the label of the nearest center
### Return : np array of size N
def update_labels(distances):
    
    u_label=np.argmin(distances,axis=1)       
    return u_label
u_label=update_labels(distance)
print(u_label)
```

```python
### TODO 4 : M step
### Update the centers to the mean of the data points assigned to it
### Return : np array of size Kx2
def update_centers(data, labels, K):
    newcenters=np.zeros((K,2))
    for k in range(K):
        points=data[labels==k]
        # print(points)
        if len(points)>0:
            newcenters[k]=np.mean(points,axis=0)
    return newcenters
K=2
newcenters=update_centers(data,u_label,K)
print(newcenters)
```

```python
### TODO 6 : Check convergence
### Check if the labels have changed from the previous iteration
### Return : True / False
def check_termination(labels1, labels2):
    x=np.array_equal(labels1,labels2)             #element wise equality 
    
    return x
check=check_termination(label,u_label)
print(check)
```

```python
### DON'T CHANGE ANYTHING IN THE FOLLOWING FUNCTION
def kmeans(data_path:str, K:int, init_centers):
    '''
    Input :
        data (type str): path to the file containing the data
        K (type int): number of clusters
        init_centers (type numpy.ndarray): initial centers. shape = (K, 2) or None
    Output :
        centers (type numpy.ndarray): final centers. shape = (K, 2)
        labels (type numpy.ndarray): label of each data point. shape = (N,)
        time (type float): time taken by the algorithm to converge in seconds
    N is the number of data points each of shape (2,)
    '''
    data = load_data(data_path)    
    centers = initialise_centers(data, K, init_centers)
    labels = initialise_labels(data)

    start_time = time.time() # Time stamp 

    while True:
        distances = calculate_distances(data, centers)
        labels_new = update_labels(distances)
        centers = update_centers(data, labels_new, K)
        if check_termination(labels, labels_new): break
        else: labels = labels_new
 
    end_time = time.time() # Time stamp after the algorithm ends
    return centers, labels, end_time - start_time 
```

```python
### TODO 7
def visualise(data_path, labels, centers):
    data = load_data(data_path)

    # Scatter plot of the data points
    plt.scatter(data[:, 0], data[:, 1], c=labels, s=50, cmap='viridis')
    plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

    ### Set title as 'K-means clustering'
    plt.title("K-means clustering")
    ### Set xlabel as 'Longitude'
    plt.xlabel("Longitude")
    ### Set ylabel as 'Latitude'
    plt.ylabel("Latitude")
    ### Save the plot as 'kmeans.png'
    plt.savefig('kmeans.png')

    ## DO NOT CHANGE THE FOLLOWING LINE
    return plt
```

```python
### After you have completed the above functions, run the following code to generate the plot
data_path = 'spice_locations.txt'
K, init_centers = 4, None
centers, labels, time_taken = kmeans(data_path, K, init_centers)
print('Time taken for the algorithm to converge:', time_taken)
visualise(data_path, labels, centers)
```
